```python
import numpy as np
import pandas as pd
from pathlib import Path
import datetime as dt
import plotly.express as px
```

```python
Path.cwd()
```

```python
file_path_stock_price = Path(r'C:\Users\paulk\project-1\python\stock_price_data\stock_price_data.csv')
```

```python
stock_prices = pd.read_csv(file_path_stock_price)

stock_prices_plot = pd.melt(stock_prices,
                            id_vars=['Period'],
                            value_vars=['Alaska Air Group Inc Price', 'American Airlines Group Inc Price', 'Delta Air Lines Inc Price', 'JetBlue Airways Corp Price', 'Southwest Airlines Co Price', 'United Airlines Holdings Inc Price'])
```

```python
fig = px.line(stock_prices_plot, x='Period', y='value', color='variable', title='Airline Stock Prices(USD) from August 11 - 14 (2020)',
             labels={"value": "Price (USD)", "variable": "Stock"},
             width=800, height=500)

fig.show()
```

```python
stock_prices2 = stock_prices.set_index(['Period'])
pct_change = stock_prices2.pct_change()
pct_change.dropna(inplace=True)
```

```python
pct_change_fixed = pct_change.reset_index()
```

```python
pct_change_plot = pd.melt(pct_change_fixed,
                            id_vars=['Period'],
                            value_vars=['Alaska Air Group Inc Price', 'American Airlines Group Inc Price', 'Delta Air Lines Inc Price', 'JetBlue Airways Corp Price', 'Southwest Airlines Co Price', 'United Airlines Holdings Inc Price'])
```

```python
fig = px.line(pct_change_plot, x='Period', y='value', color='variable', title='Airline Stock Price %change from August 11 - 14 (2020)',
             labels={"value": "% Change", "variable": "Stock"},
             width=1000, height=600)

fig.show()
```
